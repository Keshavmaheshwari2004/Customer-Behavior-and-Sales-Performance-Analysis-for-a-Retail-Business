```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\DELL\Downloads\SampleSuperstore.csv", encoding='latin1')
print(df)
```

# Initial Data Inspection

```python
print(df.head(5))         # View first few rows
print(df.info())         # Check data types & non-null values
print(df.describe())     # Summary stats
```

 # Detect Missing Values

```python
print(df.isnull().sum())    # Count missing values per column
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')  # Visualize missing data
plt.title("Missing Data Heatmap")
plt.show()
```

# Detect & Remove Duplicates

```python
print("Duplicates: ", df.duplicated().sum())   # Count duplicates
df = df.drop_duplicates()                      # Remove duplicates
```

```python
df.to_csv("cleaned_data.csv", index=False)
print(df)
```
